В групі рухів правильного восьмикутника задайте циклічну підгрупу порядку 4. Побудуйте для неї ліві і праві класи суміжності та перевірте чи буде ця група нормальною. 

```python
D = DihedralGroup(8)
subgroups = D.subgroups()
cycle_sub = []
for i in subgroups:
        if i.order() == 4 and i.is_cyclic():
            cycle_sub.append(i)
print(cycle_sub)
a = D("(1,3,5,7)(2,4,6,8)")
H = D.subgroup([a])
tr = D.cosets(H, side='right')
tl = D.cosets(H, side='left')
if tr == tl:
    print("The group is normal group.")
else:
    print("This group isn't normal group.")
```

## Гіпотеза
***
**D_n**

Якщо **n** — парне, то кількість нормальних підгруп дорівнює: </p>
**(к-сть дільників - 1) + 3 + 1,** </p>
де: 
* 3 — це кількість нормальних підгруп порядку n;
* 1 — це кількість нормальних підгруп порядку 2*n. 
***
Якщо **n** — непарне, то кількість нормальних підгруп дорівнює: </p>
**к-сть дільників + 1,**  </p>
де: 
* 1 — це кількість нормальних підгруп порядку 2*n.
    

```python
n = factor(470448)
n
```

```python
p = []
c = 1
for num, power in n:
    p.append(power)
for i in p:
    c *= (i+1)
res = c - 1 + 3 + 1
print(f"К-сть нормальний підгруп D_470448: {res}")
```
